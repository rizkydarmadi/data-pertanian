```python
# import data
import pandas as pd

top_provinsi = pd.read_csv('dataset/top_provinsi_2000-2021.csv')
top_provinsi
```

```python
hasil_tani = pd.read_csv('dataset/anual_value.csv')
grk = pd.read_csv('dataset/GRK.csv')
hasil_tani['total'].head(20).values
grk['Jumlah'].head(20).values
grk['Tahun'].head(20).values

df1 = pd.DataFrame({
    'hasil':hasil_tani['total'].head(20).values,
    'jumlah_gas':grk['Jumlah'].head(20).values,
    'tahun':grk['Tahun'].head(20).values
})
df1
```

```python
from sklearn import preprocessing

df2 = pd.DataFrame({
    'gas':[int(i) for i in df1['jumlah_gas'].values],
    'hasil':[int(i) for i in df1['hasil'].values]
})

scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(df2)
df3 = pd.DataFrame(x_scaled,columns=['gas','hasil'],dtype=float)
df3

```

```python
import matplotlib.pyplot as plt
plt.scatter(x=df3['hasil'],y=df3['gas'])
plt.show()
```
